```python
from ultralytics import YOLO
from tabulate import tabulate

models = {
    "YOLOv8": "/Users/arjavpatil/Documents/Review2/best.pt",
    "YOLOv9c": "/Users/arjavpatil/Documents/Review2/best (3).pt",
    "YOLOv11": "/Users/arjavpatil/Documents/Review2/best (4).pt"
}

table_data = []

for name, path in models.items():
    print(f"\n🔍 Evaluating {name}...\n")

    model = YOLO(path)

    metrics = model.val(
        data="/Users/arjavpatil/Documents/Review2/PanelDetect.v8i.yolov9/data.yaml",
        verbose=False
    )

    # 🔥 SAFE EXTRACTION (IMPORTANT)
    results_dict = metrics.results_dict

    mp = results_dict.get("metrics/precision(B)", 0)
    mr = results_dict.get("metrics/recall(B)", 0)
    map50 = results_dict.get("metrics/mAP50(B)", 0)
    map5095 = results_dict.get("metrics/mAP50-95(B)", 0)

    # F1 score
    f1 = (2 * mp * mr) / (mp + mr + 1e-6)

    # Store row
    table_data.append([
        name,
        mp,
        mr,
        map50,
        map5095,
        f1
    ])

# -------------------------------
# PRINT TABLE
# -------------------------------
headers = ["Model", "Precision", "Recall", "mAP50", "mAP50-95", "F1"]

print("\n📊 FINAL MODEL COMPARISON TABLE\n")
print(tabulate(table_data, headers=headers, floatfmt=".4f"))
```
